## 3. Create at least three graphs to help answer these questions. Ensure the graphs are clearly readable and are labeled appropriately and professionally.

```python
'''
Utilize a histogram to view the distribution of insurance charges.
Use plt.hist() to create the histogram.
'''
# Specify number of bins and calculate mean and standard deviation for best fit line.
bin_number = 25
mu = df['charges'].mean()
standard_dev = df['charges'].std()

# Generate the histogram of the data.
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
n, bins, patches = plt.hist(df['charges'], bin_number, rwidth = .97)
plt.axvline(df['charges'].median(), color='k', linestyle='--', label ='Median')
plt.axvline(df['charges'].mean(), color='k', linestyle='-', label ='mean')
plt.title('Insurance Charges Distribution')
plt.xlabel('Insurance Charges ($)')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
'''
Find the Mean, Median, and Skewness to see if the chart histogram accordingly. 
'''
print('Mean: ${}, Median: ${}, Standard Deviation:{}, and Skew:{}'.format(round(df['charges'].mean(),2),\
                                                                          round(df['charges'].median(),2),\
                                                                          round(df['charges'].std(),2),\
                                                                          round(df['charges'].skew(),4)))
```

```python
'''
Use a box plot to compare smoker vs. non-smoker charges.
Try using seaborn sns.boxplot().
'''
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
sns.boxplot(x='smoker', y='charges', data=df)
plt.xlabel('Smoker (Yes/No)')
plt.ylabel('Charges ($)')
plt.title('Insurance Charges by Smoker Status')
```

```python
'''
Find the median value of smoker and non-smoker insurance charges.
'''
round(df.groupby('smoker')['charges'].median(),2)
```

```python
'''
Use a box plot to compare insurance charges based by gender.
Try using seaborn sns.boxplot().
'''
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
sns.boxplot(x='sex', y='charges', data=df)
plt.xlabel('Gender')
plt.ylabel('Charges ($)')
plt.title('Insurance Charges by Gender')
```

```python
'''
Find the median value of male and female insurance charges.
'''
round(df.groupby('sex')['charges'].median(),2)
```

```python
'''
Try using seaborn sns.countplot() to determine the number of males/females that are smokers and non-smokers.
'''
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
sns.countplot(x='sex', data=df, hue = 'smoker')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Gender by Smoker Status')
```

```python
'''
Find the count of male and female smokers vs. non-smokers.
'''
df.groupby(['sex','smoker'])['charges'].count()
```

```python
'''
Display a correlation heatmap. Utilize sns.heatmap() to generate the figure.
'''
# Calculate the correlation coefficient with corr().
correlation_number = df.corr()

# Create the heatmap for the correlation coefficients calculated above.
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
sns.heatmap(correlation_number, annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap of US Insurance Charges Data')
```

```python
'''
Plot the age against charges on a scatter plot. This has the highest correlation based
on the heatmap above. 
Also, group out the smoker status within the plot.
'''
fig, ax = plt.subplots(1, 1, figsize=(10,7), tight_layout = True)
sns.scatterplot(x='age', y='charges', data=df, hue = 'smoker')
plt.xlabel('Age')
plt.ylabel('Charges ($)')
plt.title('Insurance Charges Against Age (by Smoker Status)')
```
